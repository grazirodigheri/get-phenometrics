```r
#
# This function requires the \CRANpkg{cropphenology} package develope by Araya et al., (2018)
# 
# install_github("SofanitAraya/CropPhenology")
#
# @title Get phenological metrics data using CropPhenology package
# 
# ------------------------------------------------------------
# CropPhenology function

library("CropPhenology")

# Define a função
extract_metrics_cropphenology <- function(serie, percent) {
    
#     if(!requireNamespace("CropPhenology", quietly = TRUE)){
#     warning("The CropPhenology package is required for this function")
#     return(NULL)
#   }
    
    # Extrai as métricas phenológicas da série temporal
    metrics <- SinglePhenology(serie, Percentage=percent)
    
    metrics_df <- data.frame(
        OnsetV = metrics[1],
        OnsetT = metrics[2],
        MaxV = metrics[3],
        MaxT = metrics[4],
        OffsetV = metrics[5],
        OffsetT = metrics[6],
        LengthGS = metrics[7]
    )
    
    return (metrics_df)
}
```
